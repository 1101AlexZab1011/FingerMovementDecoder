```python
import time
import numpy as np
import PySimpleGUI as sg
import pandas as pd

sg.theme('DarkGrey5')

layout_file_chooser = [
    [
        sg.Column(
            [[sg.T('Choose file', font='ANY 15', pad=(100, (25, 50)), key='path')]],
            justification='center'
        )
    ],
    [
        sg.Column(
            [[
                sg.FileBrowse(
                    key='select-file',
                    file_types=(("CSV Files", "*.csv"),),
                    initial_folder='/home/user/Downloads/',
                    enable_events=True,
                    target='path',
                ),
                sg.B('Submit', key='submit-button', pad=((100, 0), 0))
            ]],
            justification = 'center'
        )
    ]
]

main_layout = [
    [
        sg.Column(
            [[
                sg.T('Test on Anatomy', font='ANY 15')
            ]],
            justification = 'center'
        )
    ],
    [
        sg.Column(
            [[
                sg.T('0% done', font='Any 15', key='percentage')
            ]],
            justification = 'center'
        )
    ],
    [
        sg.Column(
            [[
                sg.T('Question should be here', font='ANY 15', key='question')
            ]],
            justification = 'center',
            pad=(0,150)
        )
    ],
    [
        sg.Column(
            [[
                sg.B('Start', font='ANY 15', key='start_button'),
                sg.B('Close', font='ANY 15', key='close-button1')
            ]],
            justification = 'center',
            key='initial',
        ),
        sg.Column(
            [[sg.Text('', key='timer',)]],
            justification = 'center',
            key='timer-col',
            visible=False
        ),
        sg.Column(
            [[
                sg.B('True', font='ANY 15', key='true_answer'),
                sg.B('False', font='ANY 15', key='false_answer'),
                sg.B('Close', font='ANY 15', key='close-button')
            ]],
            justification = 'center',
            key='main',
            visible=False
        ),
    ],
    [
        sg.Column([[
            sg.Multiline(size=(150, 15), key='wrong-answers', background_color='#eeeeee')
        ]], justification = 'center', pad=50)
    ]
]

window = sg.Window(
    title='Happy Testing', font='ANY 15',
    layout=layout_file_chooser,
)

path = None

while True:
    e, v = window()
    if e is None or e == sg.WIN_CLOSED or e == 'submit-button':
        if v['select-file']:
            path = v['select-file']
        else:
            path = None
        break

window.close()

def time_as_int():
    return int(round(time.time() * 100))

if path is not None:
    questions_csv = pd.read_csv(path)
    size = questions_csv.shape[0]
    questions = questions_csv.to_numpy()
    questions = np.take(questions, np.random.permutation(questions.shape[0]),axis=0,out=questions)


    window = sg.Window(
        title='Happy Testing', font='ANY 15',
        layout=main_layout,
        finalize=True,
        resizable=True,
    )

    i = 0
    next_question = False
    actual_answer = None
    answer = None
    correct = 0
    current_time = 0
    start_time = time_as_int()
    
    while True:
        
        e, v = window.read(timeout=10)
        
        if window['timer-col'].visible:
            current_time = time_as_int() - start_time
            window['timer'].update('{:02d}:{:02d}'.format((current_time // 100) // 60,
                                                        (current_time // 100) % 60))
        
        if e == '__TIMEOUT__':
            continue
        
        if e is None or e == sg.WIN_CLOSED or e == 'close-button' or e == 'close-button1':
            break
        
        if e == 'start_button':
            next_question = True
            window['initial'].update(visible=False)
            window['main'].update(visible=True)
            window['timer-col'].update(visible=True)
            
        if e == 'true_answer':
            answer = 'true'
            
        if e == 'false_answer':
            answer = 'false'
            
        if answer is not None and actual_answer is not None and answer[0] == actual_answer[0].lower():
            correct +=1
            window['question'](background_color='green')
            window.refresh()
            time.sleep(.5)
            window['question'](background_color='#343434')
            window.refresh()
            
        elif answer is not None and actual_answer is not None and answer[0] != actual_answer[0].lower():
            window['question'](background_color='red')
            window.refresh()
            time.sleep(.5)
            window['question'](background_color='#343434')
            window.refresh()
            window['wrong-answers'].update(f'{question} {actual_answer}\n', append=True)
            
        if next_question:
            print(current_time)
            current_time = 0
            start_time = time_as_int()
            question = questions[i][0]
            actual_answer = questions[i][1]
            window['question'].update(question)
            i+=1
            window['percentage'].update(f'{i/size*100 : .2f}% done: {i}/{size}')
            if i == questions.shape[0]:
                break
    window.close()
```
